```python
import numpy as np
import matplotlib.pyplot as plt

x_min, x_max = -0.5, 0.5
y_min, y_max = -1.2, 0.2
step = 0.01


x, y = np.meshgrid(np.arange(x_min, x_max, step),
                   np.arange(y_min, y_max, step))


eq1 = np.cos(y + 0.5) + x - 0.8
eq2 = np.sin(x) - 2*y - 1.6

fig, ax = plt.subplots(figsize=(8, 8))

ax.contour(x, y, eq1, levels=[0], colors='purple')

ax.contour(x, y, eq2, levels=[0], colors='green')

ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Графік системи рівнянь')

plt.grid(True)

ax.arrow(x_min, 0, x_max - x_min, 0,
         head_width=0.03, head_length=0.03,
         fc='black', ec='black')

ax.arrow(0, y_min, 0, y_max - y_min,
         head_width=0.03, head_length=0.05,
         fc='black', ec='black')

plt.show()
```

```python
# Розв'язок системи рівнянь методом простої ітерації
import numpy as np
from scipy import optimize
import math

x0, y0 = -0.13, -0.85

def f1(y):
    return 0.8 - math.cos(y + 0.5)

def f2(x):
    return (math.sin(x) - 1.6) / 2

def iter(x, y, e):
    xn = x
    yn = y

    xn1 = f1(y)
    yn1 = f2(x)

    n = 1

    while (abs(xn1 - xn) >= e) or (abs(yn1 - yn) >= e):
        xn = xn1
        yn = yn1

        xn1 = f1(yn)
        yn1 = f2(xn)

        n += 1

    print('Simple iteration:')
    print('x =', xn)
    print('y =', yn)
    print('The amount of iteration =', n)

iter(x0, y0, 0.001)

def f3(x):
    return (
        math.cos(x[1] + 0.5) + x[0] - 0.8,
        math.sin(x[0]) - 2*x[1] - 1.6
    )

s = optimize.root(f3, [0., 0.], method='hybr')

print('Check', s.x)
```
